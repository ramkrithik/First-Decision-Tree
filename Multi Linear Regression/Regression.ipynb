Importing the packages

```python
import numpy as np
import pandas as pd
```

Reading the files

```python
df = pd.read_csv("/content/Train.csv")
print(df.shape)
print(df.dtypes)
df.head()
```

Extracting the objects for one hot encoding

```python
df=df.drop(columns=["Hometown"])
df1 = df.select_dtypes(include=[object])
df=df.fillna(method='bfill')
df.head()
```

Encoding the columns

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df2 = df1.apply(le.fit_transform)
df2.head()
```

```python
df2=df2.drop(columns=['Employee_ID'])
df2.head()
```

Replacing the primary dataframe with the encoded values

```python
df['Gender']=df2['Gender']
df['Relationship_Status']=df2['Relationship_Status']
df['Unit']=df2['Unit']
df['Decision_skill_possess']=df2['Decision_skill_possess']
df['Compensation_and_Benefits']=df2['Compensation_and_Benefits']
df.head()
df3=df.drop(columns=['Employee_ID'])
df3.to_csv("/content/processedtrain.csv",index=False)
```

```python
lab=["Attrition_rate"]
result = df.copy()
for feature_name in lab:
      max_value = df[feature_name].max()
      min_value = df[feature_name].min()
      result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
for i in lab:
  df[i]=result[i]
```

```python
selected=["Gender","Unit","Compensation_and_Benefits","Attrition_rate"]
dfs=pd.DataFrame(columns=selected)
for i in selected:
  dfs[i]=df[i]
dfs.head()
```

Creating the arrays

```python
x=np.array(dfs.iloc[:,1:(len(dfs.columns)-1)])
y=np.array(dfs.iloc[:,(len(dfs.columns)-1)])
```

```python
from sklearn.model_selection import train_test_split
xt,xte, yt, yte = train_test_split(x, y,test_size=0.03)
xt
```

Linear Regression model

```python
from sklearn import linear_model
regress = linear_model.LinearRegression()
regress.fit (xt,yt)
print ('Coefficients: ', regress.coef_)
print ('Intercept: ',regress.intercept_)
```
